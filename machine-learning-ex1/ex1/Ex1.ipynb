#  Part 3: Gradient Descent Implementation

```python
def computeCost(X, y, theta):
    m = X.shape[0] # get number of training examples
    return (sum(np.power(np.dot(X,theta) - y, 2))/(2*m))[0,0];
```

```python
m = data.shape[0] # number of training examples

# Add a column of ones to x to accomodate the theta_0 intercept term
data.insert(0, 'Ones', 1)

# set X (training data) and y (target variable)
cols = data.shape[1] # Get number of cols
X = data.iloc[:,0:cols-1] # select all rows, columns 0(zeros) and 1(population)
y = data.iloc[:, cols-1:cols] # select all rows, column 2 (profit)

X.head()
```

```python
y.head()
```

```python
# convert from data framesto numpy matrices
X = np.matrix(X.values)
y = np.matrix(y.values)

# initiliaze fitting parameters matrix
theta = np.matrix('0;0')

X.shape, theta.shape, y.shape
```

```python
print computeCost(X,y,theta)
```

```python
def gradientDescent(X, y, theta, alpha, iterations):
    m = X.shape[0]
    J_history = np.zeros(iterations)
    
    for i in range(iterations):
        error = np.dot(X, theta) - y
        delta = np.dot(X.T, error) / m
        theta = theta - alpha * delta;
        
        J_history[i] = computeCost(X, y, theta)
        
    return theta, J_history
```

```python
# initialize variables for learning rate and iterations
alpha = 0.01
iters = 1500

# perform gradient descent
theta, cost = gradientDescent(X, y, theta, alpha, iters)

print theta, cost
```

```python
computeCost(X, y, theta)
```

```python
# Verify that gradient descent is working to see if the cost function is always decreasing then converges
fig, ax = plt.subplots(figsize=(12,8))
ax.plot(np.arange(iters), cost, 'r')
ax.set_title('Cost function over iterations')
ax.set_xlabel('iterations')
ax.set_ylabel('cost function')
```

```python
x = np.linspace(data.population.min(), data.population.max(),10)
f = theta[0,0] + (theta[1,0] * x )

fig, ax = plt.subplots(figsize=(12,8))
ax.plot(x,f,'r',label='Prediction')
ax.scatter(data.population, data.profit, label="Training Data")
ax.legend(loc=2)
ax.set_xlabel('Population')
ax.set_ylabel('Profit')
ax.set_title('Predicted Profit vs. Population Size')
```

```python
theta0_vals = np.linspace(-10, 10, 100)
theta1_vals = np.linspace(-1, 4, 100)

J_vals = np.zeros((theta0_vals.size, theta1_vals.size))

for i in range(len(theta0_vals)):
    for j in range(len(theta1_vals)):
        t = np.array([theta0_vals[i], theta1_vals[j]])
        J_vals[i,j] = computeCost(X, y, t);

J_vals = J_vals.T
```

```python
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111, projection='3d')
theta0_vals, theta1_vals = np.meshgrid(theta0_vals, theta1_vals)
ax.plot_surface(theta0_vals, theta1_vals, J_vals)
ax.set_xlabel('theta_0')
ax.set_ylabel('theta_1')
plt.show()
```

```python
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111)
ax.contour(theta0_vals, theta1_vals, J_vals, np.logspace(-2, 3, 20))
ax.set_xlabel('theta_0')
ax.set_ylabel('theta_1')
plt.show()
```
